```python
import cv2
import os
from PIL import Image
output_dir = 'DATA_BAW/Triche/'
i = 1
images = 'DATA/Triche/'
fichiers_dossier = os.listdir(images)
for img in fichiers_dossier:
  chemin_complet = os.path.join(images, img)
  image =cv2.imread(chemin_complet)
  image_noir_blanc = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  output_path = os.path.join(output_dir, f'result+{i}.jpg')
  cv2.imwrite(output_path, image_noir_blanc)
  i += 1
```

```python
from PIL import Image, ImageEnhance
import os
images_path = 'DATA_BAW/Triche/'
output_directory = 'DATA_BAW/Triche/'
fichiers_dossier = os.listdir(images_path)
os.makedirs(output_directory, exist_ok=True)
gamma_factors = [0.5, 1.2,2.0]
i=0
for image in fichiers_dossier:
    print(i)
    complet=Image.open((os.path.join(output_directory,image)))
    print(chemin_complet)
    for factor in gamma_factors:
       enhancer = ImageEnhance.Brightness(complet)
       gamma_corrected_image = enhancer.enhance(factor)
       output_path = os.path.join(output_directory, f'gamma_corrected_{i}.jpg')
       gamma_corrected_image.save(output_path)
       i+=1
```
